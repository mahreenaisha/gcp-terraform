```python
// ---------------------------------------------------------
// Terraform block — defines settings and required providers
// ---------------------------------------------------------
terraform {
  required_providers {
    // We are telling Terraform that we'll use the "google" provider,
    // which comes from the "hashicorp" namespace in the Terraform registry.
    google = {
      source  = "hashicorp/google"
      
      // The version constraint "~> 7.8.0" means:
      // Use version 7.8.0 or any newer patch version (e.g., 7.8.1, 7.8.2),
      // but not 7.9.0 or higher (no major/minor jumps).
      version = "~> 7.8.0"
    }
  }
}

// ---------------------------------------------------------
// Provider block — configures how Terraform connects to Google Cloud
// ---------------------------------------------------------
provider "google" {
  // The project ID in Google Cloud where Terraform will create resources.
  project = "terraform-project"

  // The default region for resources that require a region.
  // Some resources (like Compute Engine VMs) need a specific region.
  region  = "us-central1"
}

// ---------------------------------------------------------
// Resource block — defines what you want to create in GCP
// ---------------------------------------------------------
resource "google_storage_bucket" "my-bucket" {
  // A human-readable name for the bucket (must be globally unique across all of GCP).
  name          = "vit-project-githubdemo-bucket"

  // The geographic location of the bucket.
  // "US" is a multi-region — your data is stored redundantly across multiple data centers in the U.S.
  location      = "US"

  // If true, Terraform will delete all objects in the bucket when you destroy it.
  // Useful for cleanup during testing or development.
  force_destroy = true

  // Enforces a policy that prevents the bucket from being made public.
  // Helps ensure your data isn't accidentally exposed.
  public_access_prevention = "enforced"
}
```
